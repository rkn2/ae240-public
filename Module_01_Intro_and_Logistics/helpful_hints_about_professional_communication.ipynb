**Proofreading Email Checklist:**

* **Greeting:**
   - **Description:** Do you begin with a formal greeting or salutation to address the recipient appropriately.
   - **Good Example:** "Dear Professor Smith," "Hello Dr. Johnson"
   - **Bad Example:** "hey"

*  **Grammar and Spelling:**
   - **Description:** Use built in spell check for grammatical errors and ensure proper spelling throughout the email.
   - **Good Example:** "*I* am writing *to* request an extension."
   - **Bad Example:** "*i* am writing *too* request an extention."
  - **Good Example 2:** "*Please* provide the necessary information by Friday."
   - **Bad Example 2:** "*Plz* provide the neccessary info by Friday."

* **Punctuation:**
   - **Description:** Do I use all punctuation marks correctly? Do I have all the punctuation I need?
   - **Good Example:** "Thank you for your help."
   - **Bad Example:** "thnk u for your hlp"

* **Sentence Structure:**
   - **Description:** Make sure you review your sentence structure for clarity and correctness. If you said it aloud would it sound right?
   - **Good Example:** "The report you requested is attached."
   - **Bad Example:** "attached the report you requested."

* **Clarity and Cohesion:**
   - **Description:** Confirm that the email's content flows logically and cohesively, using appropriate transitional words. If you said it aloud would it sound right?
   - **Good Example:** "Additionally, I wanted to discuss the upcoming meeting."
   - **Bad Example:** "Meeting tomorrow. Also, I need the report."
