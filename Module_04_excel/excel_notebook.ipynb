## **Using formulae**

First, we need to calculate the **stress** on the concrete. Stress is defined as the force applied per unit area:

$$ \sigma = \frac{F}{A} $$

Where:
- $\sigma$ (sigma) is the stress
- $F$ is the force
- $A$ is the cross-sectional area of the concrete cylinder

Our concrete cylinder has a diameter of 4 inches. The area is given by:

$$ A = \pi r^2 $$

1. In your spreadsheet, first calculate the area of the cylinder in a separate cell. Remember to use the radius, not the diameter!

2. Create a new column for **Stress (psi)**.

3. In the first cell of the new column, enter the formula to calculate the stress. Divide the first force value by the area you calculated. **Important:** When you drag this formula down, you want to keep referencing the same area cell. To do this, use a $ before the letter and number of the cell reference for the area (e.g., `$A$1`). This is called an absolute reference. For more info on dollar signs in excel: https://dedicatedexcel.com/dollar-signs-in-excel-formula/


4. Drag the formula down for all the force values.
  **Check your work:** For a force of 25,000 lbs, the stress should be approximately 2000 psi.

# **How can we examine this data closer?**
We can use a regression line to find the Modulus of Elasticity. This is a very important material property for engineers! The slope of the linear portion of the stress-strain curve is the Modulus of Elasticity.

**Creating a regression line**

**Important:** The Modulus of Elasticity should only be calculated from the initial, linear portion of the data (where the material behaves elastically).

1. Create a new scatter plot using only the data points that form the straight line at the beginning of your test (approximately the first 7-8 data points).
Right-click on your plotted data and select "Add Trendline".

2. Choose a "Linear" trendline.

3. Check the boxes for "Display Equation on chart" and "Display R-squared value on chart".

  * Do we want to force our intercept to be 0? In theory, the line should pass through 0,0 since if there is not strain on the material there should not be any stress. But real-world measurements are never perfect! There is always a bit of noise or error. The testing machine might not have been perfectly zeroed before the test, or the sensors might have a slight electronic offset. A standard linear regression calculates the line of best fit for the data you actually have. This line will naturally have a y-intercept that is very, very close to zero if the data is good. This small, non-zero intercept actually accounts for that slight, systematic measurement error.

4. The slope of this line (the 'm' in y = mx + b) is the Modulus of Elasticity for your concrete sample!



image.png

**But what happens if the equation is not displaying the full slope?**

Excel defaults to scientific notation (like 2E+06 which means 2 x 10⁶, or 2,000,000) for very large numbers to save space. Here’s how to force it to show the full number:

* Right-click directly on the trendline equation text box on your chart.

* From the menu that appears, select "Format Trendline Label...".

* A "Format Trendline Label" pane will open on the right side of your screen.

* Expand the "Number" category at the bottom of this pane.

* Change the "Category" from "Scientific" to "Number".

* You can then set the number of "Decimal places" you'd like to display. For a large number like the Modulus of Elasticity, 0 or 1 decimal place is usually fine.

You will want to store this slope and y intercept in a cell for future use!

image.png

**Evaluating the fit of our model**
Now that we have a linear model for the elastic region, we need to see how well it fits our actual data. We do this by calculating the residuals. A residual is the difference between the observed value (your actual stress data) and the predicted value from your trendline equation.

**Calculating the Residuals**
1. Create a new column in your spreadsheet called Residuals (psi).

2. For each data point in the linear section, calculate the residual using this formula: Residual = Actual Stress - (slope * Strain + intercept)

  * Use the slope and intercept values from the trendline equation you generated.

  * Make sure to use absolute references ($) for the slope and intercept cells so you can drag the formula down.

3. You should now have a column of small numbers (some positive, some negative) which represent the "error" for each point.



**What is the standard deviation of the residuals?**
Now, we'll find the standard deviation of the residuals. This tells us, on average, how far our data points deviate from the linear model.

1. Pick a blank cell in your workbook.

2. Type `=STDEV(` and then select the entire range of your calculated residual values.

3. Press Enter.

This value gives you an idea of how well your linear model fits the data. A smaller number indicates a better fit.

**Creating error bars**

Error bars visually represent the uncertainty in your data. We will use the standard deviation of the residuals to create them.

1. Click on your data series in the linear chart.

2. Go to "Chart Design" > "Add Chart Element" > "Error Bars".

3. Choose "More Error Bar Options".

4. For the vertical error bars, select "Custom" and "Specify Value".

5. For both the "Positive Error Value" and "Negative Error Value", select the cell where you calculated the standard deviation of the residuals.
