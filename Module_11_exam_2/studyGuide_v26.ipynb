# Exam 2 Study Guide:

## Conditionals, Loops, and Advanced DataFrames
Students should focus on mastering the application of conditional logic and control flow within the context of data manipulation and analysis using the Pandas and Matplotlib libraries.

## I. Core Topics and Conceptual Knowledge
Students must understand and be prepared to explain the following concepts, particularly as they apply to architectural engineering data analysis:

### A. Control Flow and Loops
1. **Conditionals (if, elif, else):**
   * Understand the fundamental structure of an if statement for making decisions based on conditions.
   * Know how to use comparison operators: greater than (`>`), less than (`<`), equal to (`==`), not equal to (`!=`), greater than or equal to (`>=`), and less than or equal to (`<=`).
   * ** Difference between if-elif-else and separate if statements:** Know that if-elif-else executes only the code block for the first true condition and skips the rest (mutually exclusive), preventing incorrect overwriting of results, which occurs with separate sequential if statements.
   * Be able to implement categorization logic using if-elif-else based on multiple thresholds (e.g., rating categorization for 'Excellent', 'Good', 'Average', 'Poor').

2. **Loops (for and while):**
   * **for loops:** Understand how to iterate over sequences (like lists or ranges) or over DataFrame rows (using `.iterrows()`).
   * **while loops:** Understand that a while loop executes as long as a condition remains True.
   * **Iterative Design:** Understand why a while loop is the appropriate choice for simulations where the number of iterations required to meet a target condition is unknown (e.g., finding the minimum safe beam height).
   * **Control Flow within Loops:** Be familiar with `continue` (to skip the current iteration) and `break` (to exit the loop entirely).

### B. Pandas Data Manipulation
3. **Data Loading and Initialization:**
   * Know how to initialize a DataFrame from a Python dictionary that I give you in the code cell.
   * Be prepared to handle data loading from external sources (CSV files) using `pd.read_csv()`.
   * Understand the purpose of importing specific libraries like `pandas` (as `pd`), `numpy` (as `np`), and `matplotlib.pyplot` (as `plt`).

4. **Data Cleaning:**
   * **Missing Values:** Be able to replace non-standard missing value placeholders (like 'NA', blanks, or dashes) with recognizable NaN values (`np.nan` or `pd.NA`) using the `.replace()` method.
   * Understand the methods for handling missing data: `.dropna()` (removing data, risking bias/loss of statistical power) or `.fillna()` (imputation).
   * **Data Types:** Know how to identify column data types using `.dtypes` and correct them (e.g., converting an object type column containing numbers to float or int) using `.astype()`.

5. **Data Filtering and Targeting:**
   * Filter rows using Boolean Indexing (`df[df['column'] condition]`).
   * Combine multiple filtering conditions using the `&` (AND) operator (for rows satisfying all conditions).
   * Use the `|` (OR) operator (for rows satisfying at least one condition).

6. **Grouped Analysis:**
   * Understand the process of grouping data using `.groupby()` (Splitting, Applying a function, Combining results).
   * Know how to calculate aggregated metrics like `.mean()`, `.std()`, and `.value_counts()` on grouped data.
   * Understand that calculating variability (like standard deviation) requires more than one data point; if a group has only one entry, its standard deviation is `NaN` (or 0 after filling).

### C. Data Visualization and Interpretation
1. Know the purpose and creation of Bar Charts (comparing categorical counts/means) and Histograms (showing distribution of continuous data).
2. Ensure all plots include a clear Title, X-axis Label, and Y-axis Label.
3. Be able to create a bar chart showing the mean of categories and adding error bars that represent the standard deviation using the `yerr` parameter.
4. Be able to describe the shape of a histogram distribution (e.g., skewed right/left, symmetric) and explain what that shape signifies about the underlying data (e.g., many low-cost projects and a few high-cost projects).
