# Analyzing Tornado Damage to the Built Environment: A Tale of Two Cities

##Project Overview
In this multi-week project, you will act as a data scientist to analyze and visualize data from two recent, devastating tornado events: the March 2020 tornado in Nashville, Tennessee, and the December 2021 tornado in Mayfield, Kentucky. Using your programming and data analysis skills, you will explore the provided datasets to uncover insights into how different types of buildings withstand or are damaged by tornadoes.

This project is designed to be a hands-on learning experience that will take you through the entire data science pipeline—from data cleaning and exploratory data analysis to creating insightful visualizations and building predictive models.

## Learning Objectives
Upon successful completion of this project, you will be able to:

* Clean and preprocess real-world, messy datasets.

* Perform exploratory data analysis (EDA) to summarize the main characteristics of a dataset.

* Create a variety of data visualizations to communicate your findings effectively.

* Formulate and answer data-driven questions.

* Implement and evaluate a basic machine learning model.

* Structure and present your findings in a clear and compelling manner.

## Project Timeline & Tasks (Suggested 4-Week Pace)

### Week 1: Data Wrangling and Initial Exploration

Your first week is focused on getting to know your datasets.

* Load Data: Load the Nashville and Mayfield datasets into your programming environment.

* Inspect Data: Check for missing values in each column. Understand the data types (integer, string, etc.) of each column.

* Data Cleaning: Develop a strategy for handling missing data (e.g., removing rows, filling with a mean or median). Correct any obvious inconsistencies you find.

* Summary Statistics: Calculate and interpret summary statistics for key numerical columns (e.g., year_built_u, building_area_m2).

* Initial Visualizations: Create basic plots to see the data distributions. A histogram of building ages (year_built_u) or a bar chart of construction types (construction_type_u) for each city would be a great start!

### Week 2: Exploratory Data Analysis (EDA)
Now you will dig deeper to uncover relationships in the data.

* Compare and Contrast: Compare the building characteristics of Nashville and Mayfield. Are there significant differences in construction types, building ages, or roof shapes (roof_shape_u) between the two cities? Use visualizations to support your comparison.

* Investigate Damage Relationships: Explore the relationship between building characteristics and the level of damage (damage_level_u).Here are a few ways you could do this:

    * Crosstabulations: Create a table (also known as a contingency table) to see the counts of each damage level for each construction_type_u. You can then visualize this with a stacked or grouped bar chart.

    * Box Plots: For numerical columns like year_built_u, create box plots to see if the distribution of building ages differs across the various damage levels.

    * Faceted Plots: Create a grid of plots (facets) to explore multi-variable relationships. For example, you could create a series of bar charts showing damage_level_u for each construction_type_u, with separate plots for each roof_shape_u.

* Formulate Questions: Based on your initial findings, develop at least three interesting questions you want to answer with the data.

### Weeks 3 & 4: Predictive Modeling and Final Report
Now it's time to use your EDA to build a predictive model!  Build a model to predict the damage_level_u based on building characteristics such as year_built_u, construction_type_u, roof_shape_u, and building_area_m2.

Split your data into a training set and a testing set.

How accurately can you predict the damage level? Which building features are the most important predictors of damage? You can start with a simple model like Logistic Regression or a Decision Tree.
