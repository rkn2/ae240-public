## Introduction to Pandas in Python

Pandas is an open-source data manipulation and analysis library for Python. It provides easy-to-use data structures and data analysis tools for handling and processing structured data. The two primary data structures in pandas are the Series and DataFrame.

### DataFrames

A DataFrame is a two-dimensional table-like data structure, where each column can have a different data type (e.g., integer, float, string). The DataFrame can be thought of as a spreadsheet or a SQL table. Each row in the DataFrame represents an observation, and each column represents a variable. DataFrames are the most commonly used data structure in pandas, and they provide a powerful way to analyze and manipulate data.

Here's the basic syntax of the rest of the `read_csv()` function:

The `read_csv()` function is the most commonly used method for reading in data from a CSV file.

https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

It has several parameters that allow you to customize how the data is read in:

*   **filepath\_or\_buffer**: The path to the CSV file or a buffer containing the CSV data.
*   **sep**: The delimiter used in the CSV file. The default is a comma (',').
*   **header**: Specifies which row of the CSV file contains the column names. The default is 'infer', which uses the first row as the column names. If set to `None`, pandas will assign default column names.
*   **names**: An array of column names to use instead of the names provided in the CSV file.
*   **index\_col**: The column to use as the index for the DataFrame. This can be a column name or a column number.
*   **usecols**: A list of columns to include in the DataFrame. You can either provide a list of column names or a list of column numbers.
*   **dtype**: A dictionary that specifies the data type of each column in the CSV file.
*   **na\_values**: A list of values to be treated as missing values.
*   **parse\_dates**: A list of columns to parse as dates. You can either provide a list of column names or a list of column numbers.
*   **infer\_datetime\_format**: If **True** and `parse_dates` is enabled, pandas will attempt to infer the format of the dates automatically.
*   **skiprows**: The number of rows to skip before reading in the data. This can be useful if the CSV file contains metadata or comments at the beginning of the file.
*   **skipfooter**: The number of rows to skip at the end of the file.

#### Selecting Specific Columns by Index (IN BRIEF)
To access columns by index in a pandas dataframe, we can use the iloc method. This method allows us to select rows and columns by their integer index positions. The general syntax for selecting columns using iloc is:


```
df.iloc[:, column_index]

```

where : selects all rows, and column_index specifies the index of the column we want to select. Note that the column index starts at 0.

To select multiple columns by index, we can pass a list of column indices instead of a single index:



```
df.iloc[:, [column_index_1, column_index_2, ...]]

```


````python
# add code here
# grab just the first column but all rows
print(df.iloc[:, 0])

print('next task.....')
# grab the second column but all rows and print to screen
print(df.iloc[:, 1])
````


It's important to keep in mind that selecting columns by index can be **error-prone**, especially if the column order changes. It's usually better to select columns by their column names, which is more flexible and less error-prone.

#### Grouping Data by a Specific Column

You can group data in a DataFrame by a specific column using the `groupby()` method. The `groupby` method in Python allows you to group a DataFrame by one or more columns, and then apply a function to the groups. It works by splitting the DataFrame into groups based on the specified column or columns, and then applying a function to each group. The result is a new DataFrame with the groups as the index and the result of the function applied to each group as the values.

```
# Group data by a specific column
grouped = df.groupby('column_name')

# Calculate the mean of each group
grouped.mean()
```

````python
# group the data by Pollutant and calculate the mean value of each numeric column
grouped = df.groupby('Pollutant').mean()
print(grouped)
````

````python
grouped = df.groupby('Pollutant').mean(numeric_only=True)
print(grouped)
````

This code will group the data by the "**Pollutant**" column and calculate the mean value of each **numeric** column for each group. The resulting output will show the mean values for each Pollutant.

Note that you can also use other aggregate functions like `sum`, `count`, `min`, `max`, etc. with `groupby` to calculate other summary statistics for each group.
