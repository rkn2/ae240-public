# Developing Engineering Judgment In General:)
In your classes, you learn the fundamental laws of science and the calculation methods. You learn how to find the right answer. Engineering judgment is the skill that helps you determine if the right answer is the right solution. :) There is a difference!

It's the intuition that tells you a calculated beam size seems too small, that a proposed HVAC system is overkill for a small office, or that a construction schedule is unrealistic. It’s the art of making sound decisions in the absence of complete data.

##What is Engineering Judgment?
Engineering Judgment is the ability to bridge the gap between theoretical calculations and real-world application. It involves a blend of:
* Technical Knowledge: The formulas, codes, and principles you learn in class.
* Practical Experience: Understanding how things are actually built, installed, and operated.
Contextual Awareness: Recognizing the influence of budget, schedule, client needs, and site constraints.
* Risk Assessment: Intuitively weighing the consequences of a potential failure.

##How to Develop It:
You can't learn engineering judgment from a textbook, but you can actively cultivate it!
1. Build Your Baseline Knowledge: Your intuition needs a foundation to stand on. You must have a mental library of "rules of thumb," common values, and key metrics. This is your sanity check. When a calculation spits out a number, you should be able to compare it to your baseline and immediately know if it's in the right ballpark. This is something you will develop over your time here at PSU! As you start to see these, write them down somewhere so you can have them to access later!

2. Be Insatiably Curious!!
* The "Why" Questions (Understanding the Origins)
  * Why does this code provision exist? What failure or problem is it trying to prevent?
  * What are the fundamental assumptions behind this formula? In what real-world scenarios would those assumptions break down?
  * Why is this the "standard practice"? What other options exist, and why is this one preferred?
* The "What If" Questions (Exploring Boundaries)
  * What happens if we double the load/demand? Does the system response scale linearly, or does it fail?
  * What is the most common mode of failure for this system/component?
  * What is the most sensitive variable in this equation? (i.e., A small change in which input causes the biggest change in the output?)
* The "How" Questions (Connecting to Reality)
  * How is this component actually constructed or installed on a job site? What are the challenges?
  * How would an architect's decision (e.g., adding a large window, creating a tall atrium) impact this calculation?
  * How does the cost of this system compare to its alternatives? When does the more expensive option become the right choice?
  * How is this system maintained, and what is its expected lifespan?

3. Connect with the Real World!: You cannot develop this skill in a vacuum. You must connect your classroom knowledge to the physical world!!
  * Get Hands-On: Work on a construction site, volunteer for Habitat for Humanity, or even just watch construction projects in your town. Pay attention to the sequence of operations.
  * Look at Existing Buildings: When you walk into a building, look up. Where are the sprinkler heads? How is the air distributed? Can you see the structural system?
  * Learn from Failure: Study case histories of structural collapses, HVAC system failures, and construction project disasters. Understanding how things go wrong is one of the fastest ways to learn how to make them go right.

Engineering judgment is a muscle! :)  The more you use it by asking critical questions and seeking out real-world context, the stronger it will become.


# Developing Engineering Judgment for Data Analysis

So hopefully at this point you have learned that engineering judgment is the skill that bridges the gap between classroom theory and real-world application. This is just as true for data analysis as it is for designing a beam or an HVAC system. The code and the statistical methods are your tools, but your engineering judgment is what tells you if the results are meaningful!!

This part of the guide will help you develop "data judgment"—the ability to look at a dataset from a building and know if it makes sense, and how to choose the right tools to clean it.

## The Most Important Question: "Does This Make Sense?"
Before you run a single line of code, and after you get every output, you must stop and ask yourself: "Does this number make sense in the context of a real building?"

Your baseline knowledge is your first line of defense!! If a sensor reports that the temperature in an office is 80°C (176°F), your engineering judgment should immediately tell you the sensor is broken. The data is physically impossible. This is the core of data judgment: comparing the numbers on the screen to the physical reality you know.

##Choosing Your Outlier Tool: A Judgment Call
In our lessons, we saw that different outlier methods can give different results. Choosing the right one is an engineering decision.

###When to Use a Fixed Threshold
Use a fixed threshold when your limit is defined by a real-world, external rule, not by the data itself.
This method is about compliance, safety, and performance targets. You are not asking "is this data point unusual compared to its peers?"; you are asking "does this data point break a specific rule?"

Ask yourself: Is there a...
  * Safety Limit? (e.g., "The pressure in this pipe must never exceed 150 psi.")
  * Building Code Requirement? (e.g., "ASHRAE 62.1 requires we provide at least 20 CFM of outdoor air per person.")
  * Performance Contract or Goal? (e.g., "The client's goal is for this building to use no more than 35 kBTU per square foot per year.")
  * Physical Impossibility? (e.g., "A relative humidity sensor cannot read 110%.")

If the answer is yes, a fixed threshold is the correct and most defensible choice.
###When to Use Standard Deviation or Z-Score
Use standard deviation or Z-score when you are looking for what is "unusual" relative to the other data points in your group.
This method is about comparative analysis and anomaly detection. You don't have an external rule, so you are letting the data itself define what is "normal."
Ask yourself: Is my goal to...
  * Find the "Energy Hog"? (Find which of the 50 office buildings in our portfolio uses the most energy compared to its peers.)
  * Perform Predictive Maintenance? (Find the one fan motor out of 100 that is drawing more current than all the others.)
  * Identify Faulty Sensors? (Find the 5 temperature sensors out of 1000 that are reading significantly differently from the sensors located right next to them.)

In these cases, you are looking for a data point that has broken the pattern of its own group. This is where Z-score and standard deviation are the right tools.
####Choosing Your Sigma (σ) Level: A Risk-Based Decision
When you use a standard deviation or Z-score approach, selecting the sigma level is another judgment call. Think of it as setting the sensitivity of your "unusualness" detector.
* 1 Sigma (σ): A "Performance Check" (Strict)
  * What it flags: Anything outside the central ~68% of the data.
  * When to use it: Use a 1-sigma threshold when you want a sensitive filter to flag a larger group of "under-performers" that might not be broken, but could be optimized. This is for continuous improvement and fine-tuning.
* 2 Sigma (σ): A "Maintenance Alert" (Common)
  * What it flags: Anything outside the central ~95% of the data.
  * When to use it: This is a good, standard baseline for most anomaly detection. A data point this far out is significantly different from the norm and likely warrants investigation.
* 3 Sigma (σ): A "Critical Failure" Alert (Very Strict)
  * What it flags: Anything outside the central ~99.7% of the data.
  * When to use it: Use a 3-sigma threshold for critical fault detection. A data point this far from the mean is so statistically rare that it is almost certainly a data error or a sign of a critical system failure.

Your data analysis is an extension of your engineering skills!! Use these methods not as a blind recipe, but as powerful tools to apply your judgment and extract meaningful, defensible insights from the data!!
