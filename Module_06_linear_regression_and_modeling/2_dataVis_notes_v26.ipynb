# Bar charts

image.png

A bar chart is a common chart type for graphing categorical data or data sorted into groups.

It consists of multiple rectangles aligned to a common baseline.

The length of each is proportional to the value it represents.

Our eyes are very good at comparing lengths when objects are aligned, making this graph easy to interpret; just one reason bar charts are common.

Our eyes start at the base and scan towards the end of each bar.

We measure the lengths relative to both the baseline and the other bars, so it's a straightforward process to identify the smallest or the largest bar.

We can also see the negative space between varying heights of bars to compare the incremental difference between them.

So let's see how to make one in python!

## Key Takeaways

- **Bar charts** (`ax.bar()`) compare numerical values across distinct categories (e.g., material costs across material types)
- **Histograms** (`ax.hist()`) show the frequency and distribution of a single continuous dataset; how values are grouped
- **Box plots** (`ax.boxplot()`) compare distributions across multiple datasets side by side; revealing median, spread, and consistency
- Always call `ax.set_xlabel()`, `ax.set_ylabel()`, and `ax.set_title()` ; labels and units are non-negotiable on engineering graphics
- When you need multiple charts in one figure, `plt.subplots(nrows, ncols)` returns an array of axes you index with `ax[row, col]`

## For the Curious: Multiple Plots in One Figure (Subplots)

You already know that `fig, ax = plt.subplots()` gives you one figure with one set of axes. If you pass `nrows` and `ncols`, you get a **grid** of axes instead; useful for showing multiple charts side by side.

The big difference: `ax` is now a 2D array, so you pick which subplot to draw on with index notation; `ax[0, 0]` for top-left, `ax[0, 1]` for top-right, etc.

Everything else (`.bar()`, `.hist()`, `.set_xlabel()`) works exactly the same as you already know.

```python
import matplotlib.pyplot as plt

cost = [75, 110, 85, 130, 150, 95, 70]  # added Ceramic Tile

# nrows=2, ncols=2 creates a 2x2 grid; ax is now an array of axes
fig_dashboard, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# top-left: bar chart
ax[0, 0].bar(materials, cost)
ax[0, 0].set_title('Cost Comparison of Materials')
ax[0, 0].set_ylabel('Cost per Square Foot ($)')
ax[0, 0].axhline(y=100, color='r', linestyle='--')

# top-right: histogram
ax[0, 1].hist(cost, bins=5, edgecolor='black')
ax[0, 1].set_title('Distribution of Material Costs')
ax[0, 1].set_xlabel('Cost per Square Foot ($)')
ax[0, 1].set_ylabel('Frequency')

# bottom-left: box plot
ax[1, 0].boxplot(data_to_plot)
ax[1, 0].set_title('Cost Distribution from Suppliers')
ax[1, 0].set_xticklabels(materials)
ax[1, 0].set_ylabel('Cost per Square Foot ($)')

# bottom-right: unused; hide it
ax[1, 1].axis('off')

plt.tight_layout()
fig_dashboard
```

# For the Super Advanced: Creating an Interactive GUI with Plotly

So far, we've created static plots with matplotlib. These are like snapshots of our data; great for reports and papers. But what if you wanted to create an interactive Graphical User Interface (GUI) or dashboard that allows you or a client to explore the data dynamically?

For this, we use a more advanced library called Plotly. Plotly creates web-based, interactive graphics where you can hover for details, zoom in on specific areas, and filter data on the fly. This is a step towards building the kind of dashboards common in the industry.

So something we could do is combine cost and embodied carbon data into a single, interactive scatter plot. This will help an engineer quickly see the trade-offs between cost and environmental impact for each material.

Plotly is an external library, so we first need to install it. Then we'll import plotly.express (as px) which is the easiest way to get started. We'll also import pandas, a powerful library for organizing data into tables called DataFrames, which is the standard input for Plotly.

Plotly is an external library (different than just an import!), so we first need to install it. Then we'll import plotly.express (as px) which is the easiest way to get started. We'll also import pandas, a powerful library for organizing data into tables called DataFrames, which is the standard input for Plotly (we will learn more about pandas after exam 1!)

```python
# Now, import the libraries we need
import plotly.express as px
import pandas as pd
```

We'll organize our data into a Pandas DataFrame and then use a single command from plotly.express to generate aninteractive plot.

```python
# Data
materials = ['Concrete', 'Steel', 'Wood', 'Brick', 'Glass', 'Aluminum']
cost = [75, 110, 85, 130, 150, 95]
embodied_carbon = [150, 450, 50, 200, 850, 1200]

# Create a Pandas DataFrame to hold all our data in a table format
df = pd.DataFrame({
    'Material': materials,
    'Cost ($/sqft)': cost,
    'Embodied Carbon (kg CO2/sqft)': embodied_carbon
})

# Create an interactive scatter plot with plotly.express
# x and y are the axes
# hover_data shows the material name when you mouse over a point
# size makes the bubble size proportional to the cost
# color assigns a unique color to each material
fig = px.scatter(df,
                 x='Cost ($/sqft)',
                 y='Embodied Carbon (kg CO2/sqft)',
                 hover_data=['Material'],
                 size='Cost ($/sqft)',
                 color='Material',
                 title='Material Cost vs. Embodied Carbon Trade-off')

# Show the interactive figure
fig.show()
```
