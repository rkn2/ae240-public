## Why a good initial guess matters

`optimize.minimize()` needs a starting guess (`x0`). For a cubic function,
which has multiple "ups and downs," a poor guess can fail entirely -- this
is worth seeing directly, even though Session 2 only mentions it in
passing.

```python
from scipy import optimize

# define an objective based on our model function
def objective(x):
  y = -model_cubic(x, params[0], params[1], params[2], params[3])
  return y

# minimize using optimize.minimize() with a poor initial guess
results = optimize.minimize(objective, 0)
print(results)
```

Ok, we saw that **we maybe got an error** for optimize.minimize() when we used the cubic fuction as our objective! This is due to poor initial guess, since the cubic function has repeated "ups and downs" throughout its domain.

```python
# minimize using optimize.minimize()
results = optimize.minimize(objective, 500)  # test different guess: 0, 500, 1000
print(results)

# find point of maximum (x and y)
xmax = results.x
ymax = model_cubic(xmax, params[0], params[1], params[2], params[3])

print("point of maximum (xmax, ymax): ")
print("xmax: ", xmax)
print("ymax: ", ymax)
```

When using a "good" initial guess, we got a "good" solution!

```python
# plotting the results

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x, model_cubic(x, params[0], params[1], params[2], params[3]), "orange")
ax.plot(xmax, ymax, 'v', markersize=16, color='green')
ax.set_ylabel('returns - profit from production ($M)')
ax.set_xlabel('n of employees')
```

So this is our point of maximum yields -- found by guessing well. Session 2 shows how the derivative finds the same point without needing to guess at all.
