## Variables

In Python, a variable is a symbolic name or identifier that represents a value or data stored in the computer's memory. Variables are used to store, manipulate, and retrieve data during the execution of a program. Think of a variable as a container that holds a value, and you can give it a meaningful name to represent what that value represents.

Key characteristics of variables in Python:

* **Assignment**: Variables are created and assigned values using the = operator. For example, `x = 10` assigns the value 10 to the variable `x`.

* **Dynamic Typing**: Python is dynamically typed, meaning you don't need to declare the data type of a variable explicitly. The interpreter determines the data type based on the assigned value. For example, `x = 10` assigns an integer to x, while `x = "Hello"` assigns a string.

* **Variable Names:** Variable names are case-sensitive and can consist of letters (a-z, A-Z), digits (0-9), and underscores (_). They must start with a letter or underscore, but *not* with a digit. For example, `my_variable`, `age`, and `_private_var` are valid variable names.

* **Reassignment:** You can change the value stored in a variable by assigning it a new value. For example, you can reassign `x` as follows: `x = 20`.

## Data Types

* **Floats**: numbers with a decimal point, also known as floating-point numbers. In Python, floats are represented by the `float` data type, like `load`.

* **Strings**: sequences of characters, which can include letters, numbers, and symbols. In Python, strings are represented by the `str` data type, enclosed in quotes, like `material`.

* **Lists**: ordered collections of objects, which can be of different data types, enclosed in square brackets. We'll meet these in just a minute, once we load our full dataset.

* **Arrays**: similar to lists, but can only contain objects of the same data type. Used for numerical computations. We'll meet these in the Linear Regression lesson.

````python
pi = 3.14
radius = diameter / 2
area = pi * radius**2

print("Radius:", radius, "inches")
print("Area:", round(area, 3), "square inches")
````

## Printing Readable Results

To print the volume above, you probably joined a string and a number together with `+`. That's called **string concatenation**. It works, but it gets clunky fast, and you have to remember to wrap numbers in `str()` first.

There are two easier ways to print a value alongside a label:
- Separate them with a **comma** in `print()`: Python adds the spacing for you, no `str()` needed.
- Use an **f-string**, which lets you embed variables directly inside `" "`. This is the prettiest option, and the only one that lets you control formatting (like rounding to a set number of decimal places).

````python
# Two equivalent ways to print a labeled value
print("Radius:", radius, "inches")       # comma approach
print(f"Radius: {radius} inches")         # f-string approach
print(f"Radius: {radius:.3f} inches")     # f-string with 3 decimal places
````

## Lists

A **list** stores multiple values in order, enclosed in `[ ]`. We just turned `load` and `strain` into lists above, exactly what a list is for: holding many related values under one name.

## Lists Are Mutable: Append and Delete

Lists can change after you create them. Two common ways:
- `list_name.append(value)` adds `value` to the **end** of the list.
- `del list_name[index]` removes the item at `index` from the list.

Let's try both on our real `load` list, then put it back the way it was.

````python
# Add an extra reading to the end of load
load.append(1000)
print(load)
print(len(load))
````

Now let's remove it again. Since we just added it to the end, `load[-1]` is it, so we can delete it by index.

````python
del load[-1]
print(load)
print(len(load))
````

`load` is back to the original 8 readings, exactly as we'll need it for everything that follows.

## For Loops

A `for` loop repeats a block of code for each item in a sequence. The indented block runs once per item.

The basic pattern looks like this:

````python
# for variable in sequence:
#     # code to run for each item in the sequence
````

* **variable**: takes on the value of each item in the sequence, one at a time, during each pass through the loop.
* **sequence**: the collection you're iterating over. It can be a list, a string, or a range of numbers, like the ones below.

````python
for reading in load:
    print(reading)
````

What if we want to print a label alongside each value?

The simplest approach is **string concatenation**: joining pieces of text with `+`. Numbers need to be converted to strings first with `str()`.

````python
for reading in load:
    print("Load reading: " + str(reading) + " lbs")
````

That works, but gets cluttered fast. **f-strings** are a cleaner way to embed variables directly inside a string: prefix the opening quote with `f` and put variable names in `{ }`.

````python
for reading in load:
    print(f"Load reading: {reading} lbs")
````

Both print the same thing. f-strings also let you control formatting, like rounding to 1 decimal place:

````python
for reading in load:
    print(f"Load reading: {reading:.1f} lbs")  # :.1f means 1 decimal place
````

You can also loop over a `range()` of numbers:

```python
for i in range(start, stop):   # generates integers from start up to (not including) stop
```

````python
for i in range(len(load)):
    print(f"Step {i}: {load[i]} lbs")
````

Let's break down how `range()` works:
* `range(start, stop, step)` is the general syntax.
* **start**: the first value in the sequence (inclusive).
* **stop**: where the sequence ends (exclusive). The sequence goes up to, but never includes, this value.
* **step** (optional): the gap between values. It defaults to `1` if you don't specify one.

So `range(0, 5)` generates `0, 1, 2, 3, 4`: it starts at 0, counts up by the default step of 1, and stops before reaching 5.

The `range()` function also accepts a **step** argument: `range(start, stop, step)`.

This is useful when you want to skip values, for example every 10 units instead of every 1.

````python
# range(start, stop, step)
# Generates: 0, 10, 20, 30, 40, 50  (stops before 60)
for depth in range(0, 60, 10):
    print(f"Depth: {depth} ft")
````

## Built-in Functions and Methods

You've already been using **built-in functions** without thinking about it. `print()`, `len()`, and `round()` are all functions that Python provides out of the box: pre-written blocks of code you call by name with `()`.

Lists have their own special functions called **methods**. A method works the same way as a function, but it belongs to a specific object and is called using a dot:


`append()` is a list method that adds a value to the **end** of the list.

````python
numbers = [10, 20, 30]
print(numbers)        # before

numbers.append(40)
print(numbers)        # after
````

## Building a New List with a For Loop

A very common pattern: start with an **empty list**, then `append()` a calculated value each iteration. This lets you apply a formula to every item in a list.

````python
# Convert load from pounds to kilonewtons (1 lb = 0.00445 kN)
load_kN = []
for f in load:
    load_kN.append(f * 0.00445)

print(load_kN)
````
