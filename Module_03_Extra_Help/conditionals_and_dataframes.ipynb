# Conditionals and dataframes

Here is a dataframe we can play with:

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Year Built': [1990, 2000, 1980, 1960, 2010],
        'Energy Usage (kWh/m^2/year)': [150, 250, 200, 100, 300]}
df = pd.DataFrame(data)
```

## for loops



### example 1

Write a program that takes a dataframe containing information about building energy usage, and calculates the age of each building in years. Use a for loop to iterate through the rows of the dataframe, and subtract the Year Built value from the current year (2023) to calculate the age of each building. Print out a statement for each building, showing its name and age.

```python
# Calculate the age of each building
current_year = 2023
for index in range(len(df)):
    age = current_year - df.loc[index, 'Year Built']
    print(f"The age of {df.loc[index, 'Building']} is {age} years.")
```

This code creates a dataframe with information about the energy usage of five buildings, and then uses a for loop to iterate through the rows of the dataframe. The loop variable index is used to index into the dataframe using the loc function. For each row, the Year Built value is subtracted from the current year (2023) to calculate the age of the building. A f-string is used to print out a statement for each building, showing its name and age.

Like always, there is more than one way to solve this problem and pandas has a functioncalled iterrows() that can be useful. Use whichever solution makes the most sense to you. 

```python
# Calculate the age of each building
current_year = 2023
for index, row in df.iterrows():
    age = current_year - row['Year Built']
    print(f"The age of {row['Building']} is {age} years.")
```

### example 2


Write a program that takes a dataframe containing information about building energy usage, and calculates the total energy usage of all buildings in kilowatt-hours (kWh). Use a for loop to iterate through the rows of the dataframe, and multiply the Energy Usage (kWh/m^2/year) value of each row by 1000 (to convert it to kWh), and add it to a running total. Print out the total energy usage of all buildings.


```python
# Calculate the total energy usage of all buildings
total_energy_usage = 0
for index in range(len(df)):
    energy_usage_kwh = df.loc[index, 'Energy Usage (kWh/m^2/year)'] * 1000
    total_energy_usage += energy_usage_kwh
print(f"The total energy usage of all buildings is {total_energy_usage} kWh.")
```

This code creates a dataframe with information about the energy usage of five buildings, and then uses a for loop to iterate through the rows of the dataframe. The loop variable index is used to index into the dataframe using the loc function. For each row, the Energy Usage (kWh/m^2/year) value is multiplied by 1000 to convert it to kilowatt-hours (kWh), and added to a running total. After the loop, the code prints out the total energy usage of all buildings.

Again however you can use iterrows if you prefer that syntax. 

```python
# Calculate the total energy usage of all buildings
total_energy_usage = 0
for index, row in df.iterrows():
    total_energy_usage += row['Energy Usage (kWh/m^2/year)'] * 1000
print(f"The total energy usage of all buildings is {total_energy_usage} kWh.")
```

## while loops

### example 1

Calculate the energy usage of each building over a given period of time: Write a program that takes a dataframe containing information about building energy usage, and calculates the energy usage of each building over a given period of time. The program should then use a while loop to iterate through the rows of the dataframe, and calculate the energy usage of each building by multiplying the Energy Usage (kWh/m^2/year) value of each row by the number of years entered by the user. The program should store the calculated energy usage in a new column of the dataframe. Finally, the program should print out a table of building names and their corresponding energy usage over the given time period.

```python
# Prompt the user to enter the number of years to calculate energy usage for
num_years = 5

# Calculate the energy usage of each building over the given period of time
i = 0
while i < len(df):
    building_energy_usage = df.loc[i, 'Energy Usage (kWh/m^2/year)'] * num_years
    df.loc[i, 'Energy Usage over Period (kWh)'] = building_energy_usage
    i += 1

# Print out a table of building names and their corresponding energy usage over the given time period
print(df[['Building', 'Energy Usage over Period (kWh)']])
```

In this example, we use a while loop to iterate through the rows of the dataframe, and calculate the energy usage of each building by multiplying the Energy Usage (kWh/m^2/year) value of each row by the number of years entered by the user. We use the loc method to access and manipulate the dataframe.

We could also use iterrows:

```python
num_years = 5
# Calculate the energy usage of each building over the given period of time
for index, row in df.iterrows():
    building_energy_usage = row['Energy Usage (kWh/m^2/year)'] * num_years
    df.at[index, 'Energy Usage over Period (kWh)'] = building_energy_usage

# Print out a table of building names and their corresponding energy usage over the given time period
print(df[['Building', 'Energy Usage over Period (kWh)']])
```

## if, elif, else statements



Calculate the energy efficiency rating of each building: Write a program that takes a dataframe containing information about building energy usage, and calculates an energy efficiency rating for each building based on its energy usage. The program should use an if, elif, else statement to assign a rating to each building based on the following criteria:

Buildings with an energy usage of less than 200 kWh/m^2/year are rated as "Highly Energy Efficient".
Buildings with an energy usage between 200 and 300 kWh/m^2/year are rated as "Moderately Energy Efficient".
Buildings with an energy usage of more than 300 kWh/m^2/year are rated as "Low Energy Efficiency".

```python
# Calculate the energy efficiency rating of each building
df['Energy Efficiency Rating'] = '' # Create an empty column to store the ratings
for i in range(len(df)):
    energy_usage = df.loc[i, 'Energy Usage (kWh/m^2/year)']
    if energy_usage < 200:
        rating = 'Highly Energy Efficient'
    elif 200 <= energy_usage <= 300:
        rating = 'Moderately Energy Efficient'
    else:
        rating = 'Low Energy Efficiency'
    df.loc[i, 'Energy Efficiency Rating'] = rating
```

In this example, we use the .loc method to access the value of each row in the Energy Usage (kWh/m^2/year) column, and use an if, elif, else statement to assign an energy efficiency rating to each building based on its energy usage. We use the .loc method to access and manipulate the dataframe within the loop.

## nested conditionals



### example 1

Write a program that takes a dataframe containing information about building energy usage, and counts the number of buildings that have an energy usage greater than 250 kWh/m^2/year. Use a for loop to iterate through the rows of the dataframe, and use a nested if statement to check if the Energy Usage (kWh/m^2/year) value of each row is greater than 250. If it is, add 1 to a running total. Print out the total number of buildings that have high energy usage.

```python
# Count the number of buildings with high energy usage
high_usage_count = 0
for energy_usage in df['Energy Usage (kWh/m^2/year)']:
    if energy_usage > 250:
        high_usage_count += 1
print(f"There are {high_usage_count} buildings with high energy usage.")
```

This code uses a for loop to iterate through the Energy Usage (kWh/m^2/year) column of the dataframe directly. For each value, a simple if statement checks if the value is greater than 250. If it is, 1 is added to a running total. After the loop, the code prints out the total number of buildings that have high energy usage.

If you want to figure out which buildings have high energy usage in the dataframe, you can modify the code to print out the name of each building that has high energy usage, like this:

```python
# Print out the names of buildings with high energy usage
high_usage = df[df['Energy Usage (kWh/m^2/year)'] > 250]
print(high_usage)
```

This code creates a dataframe with information about the energy usage of five buildings, and then uses boolean indexing to filter the dataframe to only show the buildings that have an energy usage greater than 250 kWh/m^2/year. The resulting dataframe (high_usage) contains only one row, which corresponds to Building E. You can modify the threshold value (250) to show buildings with different levels of energy usage.

Again you can do this using iterrows if you prefer:

```python
# Count the number of buildings with high energy usage
high_usage_count = 0
for index, row in df.iterrows():
    if row['Energy Usage (kWh/m^2/year)'] > 250:
        high_usage_count += 1
print(f"There are {high_usage_count} buildings with high energy usage.")
```

This code creates a dataframe with information about the energy usage of five buildings, and then uses a for loop to iterate through the rows of the dataframe. For each row, a nested if statement checks if the Energy Usage (kWh/m^2/year) value is greater than 250. If it is, 1 is added to a running total. After the loop, the code prints out the total number of buildings that have high energy usage. In this case, there is only 1 building that has high energy usage (Building E).

If you want to figure out which buildings have high energy usage in the dataframe, you can modify the code to print out the name of each building that has high energy usage, like this:



```python
# Print out the names of buildings with high energy usage
for index, row in df.iterrows():
    if row['Energy Usage (kWh/m^2/year)'] > 250:
        print(f"{row['Building']} has high energy usage.")
```

### example 2

Identify which buildings have energy usage over a certain threshold: Write a program that takes a dataframe containing information about building energy usage, and identifies which buildings have an energy usage above a certain threshold. Use a while loop to iterate through the rows of the dataframe. For each row, check if the Energy Usage (kWh/m^2/year) value is above a certain threshold (e.g. 200 kWh/m^2/year). If the value is above the threshold, print out a statement indicating that the building has high energy usage. If the value is below the threshold, continue to the next row.

```python
# Set the energy usage threshold
threshold = 200

# Identify buildings with energy usage above the threshold
i = 0
while i < len(df):
    if df['Energy Usage (kWh/m^2/year)'][i] > threshold:
        print(f"{df['Building'][i]} has high energy usage!")
    i += 1
```

In this code, we use the square bracket notation to access the Energy Usage (kWh/m^2/year) and Building columns of the dataframe. We also use an index variable i to access the individual rows of the dataframe. The while loop continues to iterate until i reaches the length of the dataframe. Inside the loop, we use a conditional statement to check if the energy usage value is above the threshold, and print out a statement if it is.

You can also do this using a function called iloc. 

```python
# Set the energy usage threshold
threshold = 200

# Identify buildings with energy usage above the threshold
i = 0
while i < len(df):
    row = df.iloc[i]
    if row['Energy Usage (kWh/m^2/year)'] > threshold:
        print(f"{row['Building']} has high energy usage!")
    i += 1
```

For each row, we check if the Energy Usage (kWh/m^2/year) value is above the threshold, and print out a statement indicating that the building has high energy usage. If the value is below the threshold, we continue to the next row.

You can modify the code by changing the threshold value to a different value, or adding additional conditional statements inside the while loop to perform more complex calculations or comparisons on the dataframe rows.
