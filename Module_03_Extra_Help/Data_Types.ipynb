# Data Types Introduction

Each data type in Python has its own set of methods and properties that can be used to manipulate and analyze the data. It's important to understand the different data types in order to effectively work with data in Python.

* Integers: Integers are whole numbers, positive or negative, without a fractional component. In Python, integers are represented by the int data type.

* Floats: Floats are numbers with a decimal point, also known as floating-point numbers. In Python, floats are represented by the float data type.

* Strings: Strings are sequences of characters, which can include letters, numbers, and symbols. In Python, strings are represented by the str data type. They are enclosed in quotes, either single or double quotes.

* Lists: Lists are ordered collections of objects, which can be of different data types. In Python, lists are represented by the list data type. They are enclosed in square brackets, and the items in the list are separated by commas.

* Arrays: Arrays are similar to lists, but they can only contain objects of the same data type. In Python, arrays are represented by the array module. They are often used for numerical computations.

* Categoricals: Categoricals are a data type used to represent data that has a finite set of possible values. In Python, categoricals are often represented using the pandas library. They are used for things like labeling data or representing discrete variables.

# Examples of each data type


## Integers

In architectural engineering, integers can be used to represent things like the number of floors in a building or the number of windows in a room. For example, if you wanted to calculate the total number of windows in a building with 10 floors and 20 windows per floor, you could use the following Python code:

```python
num_floors = 10
windows_per_floor = 20
total_windows = num_floors * windows_per_floor
print(total_windows)
```

This would output the value 200, which is the total number of windows in the building.

## Floats 

Floats are often used in architectural engineering to represent measurements of length, area, and volume. For example, if you wanted to calculate the area of a rectangular room with dimensions of 3.5 meters by 4.2 meters, you could use the following Python code:

```python
length = 3.5
width = 4.2
area = length * width
print(area)
```

This would output the value 14.7, which is the area of the room in square meters.

## Strings
Strings can be used in architectural engineering to represent things like the names of materials or the labels on a floor plan. For example, if you wanted to create a string variable to represent the name of a material, you could use the following Python code:

```python
material = "concrete"
print(material)
```

This would output the string "concrete", which represents the name of the material.

##Lists: 

Lists can be used in architectural engineering to store and manipulate data related to building components or materials. For example, if you wanted to create a list of the dimensions of windows in a building, you could use the following Python code:

```python
window_dimensions = [0.9, 1.2, 1.5, 1.8]
print(window_dimensions)
```

This would output the list [0.9, 1.2, 1.5, 1.8], which represents the dimensions of the windows.

## Arrays: 

Arrays can be used in architectural engineering to perform numerical calculations related to things like structural analysis or energy modeling. For example, if you wanted to create an array to represent the loads on a building's foundation, you could use the following Python code:

```python
import numpy as np

loads = np.array([500, 750, 1000, 1250])
print(loads)
```

This would output the array [ 500 750 1000 1250], which represents the loads on the foundation.

# How to change datatypes in dataframes

What if you have a column that is stored as one type of data but you actually want to store it as something else? 

Example 1: if something is storing as an object but you want it to be a float

```python
import numpy as np

# Create a dictionary with the building data (ignore this, I am just making a short dataframe so you can see what it looks like)
building_data = {'name': ['Building A', 'Building B', 'Building C', 'Building D'],
                 'zoning': ['R', 'R', 'C', 'C'],
                 'year_built': ['-',1920,1925, 1930], 
                 'damage_rating': [1.5,2.3,4.5, 5.0]}

# Create a pandas dataframe from the building data
building_df = pd.DataFrame(building_data)

# Print the different datatypes in the dataframe to see what 
building_df.dtypes
```

remove all the - and replace them with nans

```python
building_df['year_built'].replace('-', np.nan, inplace=True)
```

```python
# Print the different datatypes in the dataframe to see what 
building_df.dtypes
```

documentation on how to change a column in pandas to a numeric one: https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html

additional examples: https://www.w3resource.com/pandas/to_numeric.php


```python
building_df['year_built'] = pd.to_numeric(building_df['year_built'])
```

```python
# Print the different datatypes in the dataframe to see what 
building_df.dtypes
```
