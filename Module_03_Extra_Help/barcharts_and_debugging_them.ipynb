# bar charts

## general

Bar charts are a great way to represent data in a clear and easy-to-understand manner. In Python, the Matplotlib library offers a variety of tools to create bar charts, histograms, and other types of visualizations. Here's an example of how to create a simple bar chart using Matplotlib:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 24, 36, 40, 51]

plt.bar(x, y)
plt.show()
```

This code will produce a basic bar chart with five bars, representing the values in the y list. The x list provides the positions for each bar on the x-axis.

To customize the chart, you can add labels to the x and y axes, change the color and width of the bars, and add a title or legend. Here's an example that shows some of these options:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 24, 36, 40, 51]

plt.bar(x, y, color='green', width=0.4)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Bar Chart Example')
plt.legend(['Data Points'])
plt.show()
```

This code sets the color of the bars to green, and changes the width of the bars to 0.4. The xlabel, ylabel, and title functions add labels to the chart, and the legend function adds a legend to explain what the bars represent.

## AE examples

For architectural engineering, bar charts can be useful for representing data such as:

The cost of building materials
The energy consumption of different building designs
The number of hours spent on a project by each team member
The number of projects completed by each team
Here's an example of a bar chart that shows the cost of different building materials:

```python
import matplotlib.pyplot as plt

materials = ['Concrete', 'Steel', 'Wood', 'Brick']
costs = [150, 200, 100, 120]

plt.bar(materials, costs, color='blue', width=0.4)
plt.xlabel('Materials')
plt.ylabel('Cost ($)')
plt.title('Building Materials Cost Comparison')
plt.show()
```

This code creates a bar chart with four bars, representing the cost of different building materials. The xlabel, ylabel, and title functions provide labels for the chart.

Here's another example of creating a bar chart from a dataframe containing information about historic buildings rated on a scale of 0 to 5 for damage:



```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataframe with information about historic buildings and their damage rating
data = {'Building': ['Colosseum', 'Great Wall of China', 'Hagia Sophia', 'Machu Picchu', 'Pyramids of Giza'],
        'Country': ['Italy', 'China', 'Turkey', 'Peru', 'Egypt'],
        'Damage Rating': [3, 1, 4, 0, 2]}
df = pd.DataFrame(data)

# Create a bar chart of the damage rating for each building
plt.bar(df['Building'], df['Damage Rating'], color='red')
plt.xlabel('Building')
plt.xticks(rotation = 45)
plt.ylabel('Damage Rating (0-5)')
plt.title('Historic Buildings and Their Damage Rating')
plt.show()
```

# debugging with bar charts

Here is another example of creating a bar chart from a dataframe containing information about energy justice, energy prices, and affordable housing:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataframe with information about energy justice, energy prices, and affordable housing
data = {'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Philadelphia'],
        'Energy Justice Score': [3, 2, 4, 2, 3],
        'Energy Prices (cents/kWh)': [15, 16, 14, 17, 15],
        'Affordable Housing (%)': [28, 21, 25, 33, 27]}
df = pd.DataFrame(data)

# Create a bar chart of the energy justice score, energy prices, and affordable housing for each city
plt.bar(df['City'], df['Energy Justice Score'], color='green', width=0.25)
plt.bar(df['City'], df['Energy Prices (cents/kWh)'], color='blue', width=0.25)
plt.bar(df['City'], df['Affordable Housing (%)'], color='red', width=0.25)
plt.xlabel('City')
plt.ylabel('Score/Price (%)')
plt.title('Energy Justice, Energy Prices, and Affordable Housing')
plt.legend(['Energy Justice Score', 'Energy Prices (cents/kWh)', 'Affordable Housing (%)'])
plt.show()
```

This code creates a dataframe with information about five cities, including their energy justice score, energy prices, and affordable housing. Then, it creates a bar chart that includes three bars for each city representing the energy justice score, energy prices, and affordable housing. The xlabel, ylabel, and title functions provide labels for the chart, and the legend function adds a legend to explain what each bar represents.

But something went wrong! The bars are overlapping. To deal with that we can change the code to the following: 

```python
# Create a bar chart of the energy justice score, energy prices, and affordable housing for each city
plt.bar(df['City'] - 0.2, df['Energy Justice Score'], color='green', width=0.2)
plt.bar(df['City'], df['Energy Prices (cents/kWh)'], color='blue', width=0.2)
plt.bar(df['City'] + 0.2, df['Affordable Housing (%)'], color='red', width=0.2)
plt.xlabel('City')
plt.ylabel('Score/Price/Housing (%)')
plt.title('Energy Justice, Energy Prices, and Affordable Housing')
plt.legend(['Energy Justice Score', 'Energy Prices (cents/kWh)', 'Affordable Housing (%)'])
plt.xticks(df['City'])
plt.show()
```

In this updated code, we've added 0.2 to the x-axis position of the energy justice and affordable housing bars, so that they are spread out from the energy prices bars. The xticks function uses the original city names for the x-axis labels. Now all three bars can be clearly seen in the chart.
