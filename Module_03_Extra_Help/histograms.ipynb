# Histograms

Histograms are a great way to represent the distribution of a dataset in a visual manner. In Python, the Matplotlib library offers a variety of tools to create histograms and other types of visualizations.

Here's an example of how to create a simple histogram using Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30)
plt.show()
```

This code will produce a basic histogram with 30 bins, representing a dataset of 1000 random numbers generated from a normal distribution with mean 0 and standard deviation 1. The bins parameter determines the number of bins to use for the histogram.

To customize the histogram, you can add labels to the x and y axes, change the color and width of the bars, and add a title or legend. Here's an example that shows some of these options:

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, color='green', edgecolor='black', linewidth=1.2)
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Random Data Histogram')
plt.legend(['Data Points'])
plt.show()
```

This code sets the color of the bars to green, the color of the edges to black, and the width of the lines to 1.2. The xlabel, ylabel, and title functions add labels to the histogram, and the legend function adds a legend to explain what the bars represent.

## AE examples

For architectural engineering, histograms can be useful for representing data such as:

The distribution of building heights in a city
The distribution of energy usage in a building
The distribution of construction costs for a project
The distribution of team member hours spent on a project
Here's an example of a histogram that shows the distribution of building heights in a city:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataframe with information about historic buildings
data = {'Building': ['Statue of Liberty', 'Eiffel Tower', 'Leaning Tower of Pisa', 'Taj Mahal', 'Big Ben', 'Pyramids of Giza',
                     'Empire State Building', 'Burj Khalifa', 'CN Tower', 'Tokyo Skytree', 'Taipei 101', 'Shanghai Tower',
                     'Petronas Towers', 'Bank of China Tower', 'International Commerce Centre', 'Taipei 101 Tower 2', 'Abraj Al Bait Clock Tower',
                     'The Shard', 'Merdeka 118', 'Wuhan Greenland Center', 'Goldin Finance 117', 'Ryugyong Hotel'],
        'Country': ['USA', 'France', 'Italy', 'India', 'UK', 'Egypt', 'USA', 'UAE', 'Canada', 'Japan', 'Taiwan', 'China',
                    'Malaysia', 'Hong Kong', 'Hong Kong', 'Taiwan', 'Saudi Arabia', 'UK', 'Malaysia', 'China', 'China', 'North Korea'],
        'Year Built': [1886, 1889, 1173, 1632, 1859, -2560, 1931, 2010, 1976, 2012, 2004, 2015,
                       1998, 1990, 2010, 2019, 2011, 2012, 2018, 2017, 2020, -2022],
        'Height (m)': [93, 324, 56, 73, 96, 147, 443, 828, 553, 634, 509, 632, 452, 367, 484, 439, 601, 309, 630, 636, 530, 597]}
df = pd.DataFrame(data)

# Create a histogram of the heights of the historic buildings
plt.hist(df['Height (m)'], bins=5, color='orange', edgecolor='black', linewidth=1.2)
plt.xlabel('Building Height (m)')
plt.ylabel('Frequency')
plt.title('Distribution of Historic Building Heights')
plt.show()
```

This code creates a dataframe with information about historic buildings, including the world's tallest skyscraper, the Burj Khalifa, and the incomplete Merdeka 118 in Malaysia, and then creates a histogram of the height of each building. The xlabel, ylabel, and title functions provide labels for the chart. The bins parameter determines the number of bins to use for the histogram.

Here's another example of how to create a histogram from a dataframe that pertains to architectural engineering:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataframe with information about building energy usage
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Year Built': [1990, 2000, 1980, 1960, 2010],
        'Energy Usage (kWh/m^2/year)': [150, 250, 200, 100, 300]}
df = pd.DataFrame(data)

# Create a histogram of the energy usage of the buildings
plt.hist(df['Energy Usage (kWh/m^2/year)'], bins=10, color='blue', edgecolor='black', linewidth=1.2)
plt.xlabel('Energy Usage (kWh/m^2/year)')
plt.ylabel('Frequency')
plt.title('Distribution of Building Energy Usage')
plt.show()
```

This code creates a dataframe with information about the energy usage of five buildings, and then creates a histogram of the energy usage of the buildings. The xlabel, ylabel, and title functions provide labels for the chart. The bins parameter determines the number of bins to use for the histogram.

## histograms of subset data
Here is an example that takes a dataframe, sorts out a specific sample type, and then allows you to make a histogram of tensile strength for that sample type. 

```python
# just generating the data

import pandas as pd
import numpy as np

# Generate some random data for age, tensile strength, and test date
age = np.random.normal(loc=30, scale=5, size=1000)
tensile_strength = np.random.normal(loc=50, scale=10, size=1000)
test_date = pd.date_range(start='2022-01-01', end='2022-12-31', periods=1000)

# Assign sample types randomly (0, 1, 2, or 3) with equal probability
sample_type = np.random.choice([0, 1, 2, 3], size=1000)

# Create pandas dataframe
df = pd.DataFrame({
    'age': age,
    'tensile strength': tensile_strength,
    'sample type': sample_type,
    'test date': test_date
})

# Print first 5 rows of dataframe
print(df.head())
```

```python
import matplotlib.pyplot as plt
# Extract tensile strength values only for sample type 3
values_of_interest = []
for i in range(len(df)):
    if df.loc[i, 'sample type'] == 3:
        values_of_interest.append(df.loc[i, 'tensile strength'])

# Plot histogram of tensile strength values for sample type 3
yh, xh = np.histogram(values_of_interest)
xc = 0.5 * (xh[1:] + xh[:-1])
w = np.diff(xc)[0]
fig, ax = plt.subplots()
ax.bar(xc, yh, width=0.8*w)
ax.set_title('Histogram of tensile strength for sample type 3')
ax.set_xlabel('tensile strength')
ax.set_ylabel('count')
plt.show()
```
