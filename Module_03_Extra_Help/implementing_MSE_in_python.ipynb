Mean Squared Error (MSE) is a popular statistical metric used to evaluate the performance of a model, particularly in the context of regression analysis. It measures the average squared difference between the predicted and actual values of a variable.

The MSE is calculated by taking the sum of the squared differences between the predicted values and the actual values, then dividing by the total number of observations. In mathematical notation, the MSE is defined as:

MSE = 1/n * Σ(i=1 to n) (yi - ŷi)^2

where n is the total number of observations, yi is the actual value of the dependent variable for the i-th observation, and ŷi is the predicted value of the dependent variable for the i-th observation.

The MSE is a useful metric for evaluating the accuracy of a regression model, as it penalizes larger errors more heavily than smaller ones. In other words, the MSE gives greater weight to outliers or extreme values, which can provide a more complete picture of the model's performance. However, because it is calculated in terms of squared differences, the MSE can be influenced by outliers or skewed data.





### example 1

Suppose you have a dataframe that contains information about the energy usage and age of different buildings, and you want to determine if there is a relationship between the age of a building and its energy usage. 

Here is the code to write a Python program that uses scipy.stats.linregress() to perform a linear regression on the data and print out the slope, intercept, r-value, p-value, and standard error of the regression.

```python
import pandas as pd
from scipy import stats
import numpy as np

# Create a sample dataframe with information about building energy usage and age
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Year Built': [1990, 2000, 1980, 1960, 2010],
        'Energy Usage (kWh/m^2/year)': [150, 250, 200, 100, 300]}
df = pd.DataFrame(data)

# Perform a linear regression on the data
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Year Built'], df['Energy Usage (kWh/m^2/year)'])

# Print out the results
print("Slope: ", slope)
print("Intercept: ", intercept)
print("R-Value: ", r_value)
print("P-Value: ", p_value)
print("Standard Error: ", std_err)
```

I see R2 there but how can I calculate other metrics like MSE? 

Since MSE is an equation that we have above, we can calculate it (like finding the area of something using the base and height measurements). 

```python
x = df['Year Built']
y = df['Energy Usage (kWh/m^2/year)']

# Predict y values using the linear regression equation
y_pred = slope * x + intercept

# Calculate the MSE
mse = np.mean((y - y_pred) ** 2)

print(mse)
```
