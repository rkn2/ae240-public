## for loops


Basic syntax: A for loop is used to iterate over a sequence of elements (e.g., a list, tuple, or string) and perform a set of statements for each element in the sequence. The basic syntax for a for loop in Python is:

```python
for element in sequence:
    # Statements to be executed for each element
```

For example, here is a basic for loop that iterates over a list of building names and prints out each name:

```python
buildings = ['Building A', 'Building B', 'Building C']
for building in buildings:
    print(building)
```

Using range(): The range() function is often used to generate a sequence of numbers that can be used with a for loop. The basic syntax for the range() function is:

```python
range(start, stop, step)
```

For example, here is a for loop that uses the range() function to print out the first 10 even numbers:

```python
for i in range(0, 20, 2):
    print(i)
```

Nested for loops: For loops can be nested within other for loops to perform more complex iterations. For example, here is a nested for loop that iterates over two lists (building names and years) and prints out each name and year:

```python
buildings = ['Building A', 'Building B', 'Building C']
years = [1990, 2000, 1980]
for building in buildings:
    for year in years:
        print(f"{building} was built in {year}.")
```

Debugging errors: When working with for loops, it is common to encounter errors such as "list index out of range", "unhashable type: 'list'", and "indentation error". Here are some common debugging techniques for for loops:
* Check that the sequence being iterated over is of the correct type (e.g., a list).
* Check that the range being used with the for loop is of the correct size.
* Check that the indentation of the statements within the for loop is consistent.
* Use print statements to track the values of variables and pinpoint where errors are occurring.

For example, here is an example of how to debug an "unhashable type: 'list'" error:

```python
buildings = [['Building A', 1990], ['Building B', 2000], ['Building C', 1980]]
for building in buildings:
    print(building)
    # Perform some calculations on the building
```

This code will produce an "unhashable type: 'list'" error, because the 'building' variable is a list, which is a mutable type and cannot be used as a key in a dictionary or set. To fix this error, we can convert the list to a tuple (which is immutable) before using it in the loop:

```python
buildings = [['Building A', 1990], ['Building B', 2000], ['Building C', 1980]]
for building in buildings:
    building = tuple(building) # Convert the list to a tuple
    print(building)
    # Perform some calculations on the building
```

## while loops

A while loop is a type of loop in Python that is used to execute a block of code repeatedly, as long as a certain condition is true. The code inside the loop is executed repeatedly until the condition becomes false. The syntax for a while loop in Python is as follows:

```python
while condition:
    # code to be executed
```

The code inside the loop is executed as long as the condition is true. If the condition is initially false, the code inside the loop is never executed.

Example 1: Calculate the age of each building.
In this example, we can use a while loop to calculate the age of each building, given its year of construction. We can prompt the user to enter the current year, and then use a while loop to calculate the age of each building until all buildings have been processed.

```python
current_year = 2021
building_years = [1980, 1995, 2005, 1960, 2010]

i = 0
while i < len(building_years):
    age = current_year - building_years[i]
    print(age)
    i += 1
```

Example 2: Calculate the total area of all buildings.
In this example, we can use a while loop to calculate the total area of all buildings, given their individual areas. We can create a list of building areas and use a while loop to calculate the total area until all buildings have been processed.

```python
building_areas = [100, 150, 200, 80, 120]

total_area = 0
i = 0
while i < len(building_areas):
    total_area += building_areas[i]
    i += 1

print("The total area of all buildings is: ",total_area)
```

Debugging common errors

When working with while loops in Python, some common errors that can occur include infinite loops and logical errors. Here are some tips for debugging these errors:
* Use print statements to check the value of variables at different stages of the loop.
* Make sure that the condition of the while loop is properly defined and that it eventually evaluates to False.
* Check for any logical errors in the code, such as using the wrong variable names or performing incorrect calculations.
* Consider using a for loop instead of a while loop if the number of iterations is known.

By following these tips, you can identify and fix any errors in your code while using while loops in Python.

## if statements


Syntax: An if statement is used to test a condition and execute a block of code if the condition is true. The syntax of an if statement in Python is as follows:

```python
if condition:
    # code to execute if condition is true
```

The condition can be any expression that returns a boolean value (True or False).



Simple if statement example: Here's a simple example of an if statement in Python that determines if a building is taller than 100 meters:

```python
building_height = 120 # height of building in meters
if building_height > 100:
    print("This building is taller than 100 meters.")
```

In this example, the condition is building_height > 100, which is true since building_height is 120. Therefore, the code inside the if statement is executed, which prints the message "This building is taller than 100 meters."

if-else statement example: Here's an example of an if-else statement in Python that determines if a building is taller than 100 meters and prints a different message depending on the result:

```python
building_height = 80 # height of building in meters
if building_height > 100:
    print("This building is taller than 100 meters.")
else:
    print("This building is not taller than 100 meters.")
```

In this example, the condition is building_height > 100, which is false since building_height is 80. Therefore, the code inside the else statement is executed, which prints the message "This building is not taller than 100 meters."

if-elif-else statement example: Here's an example of an if-elif-else statement in Python that determines the age range of a building based on its year of construction:

```python
year_built = 1960 # year the building was constructed
current_year = 2021 # current year
age = current_year - year_built # calculate the age of the building
if age < 50:
    print("This building is less than 50 years old.")
elif age >= 50 and age < 100:
    print("This building is between 50 and 100 years old.")
else:
    print("This building is more than 100 years old.")
```

In this example, we calculate the age of a building by subtracting its year of construction from the current year. We then use an if-elif-else statement to determine the age range of the building based on its age. If the age is less than 50, the message "This building is less than 50 years old" is printed. If the age is between 50 and 100, the message "This building is between 50 and 100 years old" is printed. If the age is more than 100, the message "This building is more than 100 years old" is printed.
