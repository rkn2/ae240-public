# removing nans

## removing nans from a single column

Suppose you have a Pandas dataframe with information about the energy usage of different buildings, and some of the data is missing for certain buildings. Write a Python program that removes all NaN values from the 'Energy Usage' column of the dataframe.


```python
import pandas as pd

# Create a sample dataframe with information about building energy usage
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Energy Usage (kWh/m^2/year)': [150, 250, pd.np.nan, 100, pd.np.nan]}
df = pd.DataFrame(data)

# Remove all NaN values from the 'Energy Usage' column of the dataframe
df = df.dropna(subset=['Energy Usage (kWh/m^2/year)'])

# Print out the updated dataframe
print(df)
```

## removing nans from an entire dataframe

Suppose you have a Pandas dataframe with information about the energy usage and size of different buildings, and some of the data is missing for certain buildings. Write a Python program that removes all rows that contain NaN values from the dataframe.

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage and size
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Size (m^2)': [1000, 800, pd.np.nan, 900, 1500],
        'Energy Usage (kWh/m^2/year)': [150, pd.np.nan, 200, 100, 300]}
df = pd.DataFrame(data)

# Remove all rows that contain NaN values from the dataframe
df = df.dropna()

# Print out the updated dataframe
print(df)
```

Note that in this second example, we remove **any** rows that contain NaN values in any column.

## Remove NaNs from Specific Columns

Suppose you have a Pandas dataframe with information about the energy usage and size of different buildings, but some of the data is missing for certain buildings. However, you only want to remove the NaN values from the 'Energy Usage' and 'Size' columns. Write a Python program that removes all rows that contain NaN values in these columns only.

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage and size
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Size (m^2)': [1000, 800, pd.np.nan, 900, 1500],
        'Energy Usage (kWh/m^2/year)': [150, pd.np.nan, 200, 100, 300]}
df = pd.DataFrame(data)

# Loop over the rows of the dataframe and remove any rows that contain NaN values in the 'Size' or 'Energy Usage' columns
new_df = pd.DataFrame(columns=['Building', 'Size (m^2)', 'Energy Usage (kWh/m^2/year)'])
for index, row in df.iterrows():
    if not pd.isna(row['Size (m^2)']) and not pd.isna(row['Energy Usage (kWh/m^2/year)']):
        new_df = new_df.append(row, ignore_index=True)

# Print out the updated dataframe
print(new_df)
```

In this example, we first create a new empty dataframe with the same column names as the original dataframe. We then loop over each row of the original dataframe and check if the 'Size' and 'Energy Usage' columns contain NaN values using the pd.isna() function. If the row does not contain any NaN values in these columns, we append it to the new dataframe. Finally, we print out the updated dataframe.

While this approach works, it can be slower and less efficient than using the subset or dropna functions, especially for large dataframes. So, lets try taht: 

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage and size
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Size (m^2)': [1000, 800, pd.np.nan, 900, 1500],
        'Energy Usage (kWh/m^2/year)': [150, pd.np.nan, 200, 100, 300]}
df = pd.DataFrame(data)

# Remove all rows that contain NaN values in the 'Energy Usage' and 'Size' columns of the dataframe
df = df.dropna(subset=['Size (m^2)', 'Energy Usage (kWh/m^2/year)'])

# Print out the updated dataframe
print(df)
```

In this example, we use the dropna() function with the subset parameter to specify which columns we want to check for NaN values. We pass a list of column names to this parameter, and the function removes any rows that contain NaN values in these columns only.

Dropping NaNs from an Entire Dataframe: When using the dropna() function to remove NaN values from a dataframe, you might accidentally drop all rows in the dataframe if they all contain NaN values.

```python
import pandas as pd

# Create a sample dataframe with all NaN values
data = {'Building': [pd.np.nan, pd.np.nan, pd.np.nan],
        'Size (m^2)': [pd.np.nan, pd.np.nan, pd.np.nan],
        'Energy Usage (kWh/m^2/year)': [pd.np.nan, pd.np.nan, pd.np.nan]}
df = pd.DataFrame(data)

# Attempt to remove all NaN values from the dataframe
df = df.dropna()

# This results in an empty dataframe
print(df)
```

Dropping NaNs from a Single Column: When using the dropna() function with the subset parameter to remove NaN values from a single column, you might accidentally drop all rows in the dataframe if every value in that column is NaN.

```python
import pandas as pd

# Create a sample dataframe with a single column of all NaN values
data = {'Building': [pd.np.nan, pd.np.nan, pd.np.nan]}
df = pd.DataFrame(data)

# Attempt to remove all NaN values from the 'Building' column of the dataframe
df = df.dropna(subset=['Building'])

# This results in an empty dataframe
print(df)
```

Not Accounting for NaN Values in Calculations: When performing calculations on a dataframe that contains NaN values, you might encounter unexpected results if you do not account for the NaN values.

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Energy Usage (kWh/m^2/year)': [150, pd.np.nan, 200, pd.np.nan, 300]}
df = pd.DataFrame(data)

# Attempt to calculate the mean energy usage of all buildings
mean_energy_usage = df['Energy Usage (kWh/m^2/year)'].mean()

# This results in a NaN value, since the calculation includes the NaN values in the dataframe
print(mean_energy_usage)
```

To avoid this error, you should use the skipna parameter in calculations to exclude NaN values from the calculation. For example, df['Energy Usage (kWh/m^2/year)'].mean(skipna=True) would exclude NaN values from the calculation of the mean energy usage.

```python
   # Create a sample dataframe with information about building energy usage
   data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
           'Energy Usage (kWh/m^2/year)': [150, pd.np.nan, 200, pd.np.nan, 300]}
   df = pd.DataFrame(data)

   # Attempt to fill NaN values with a value of 0
   df.fillna(0, inplace=True)

   # This fills in NaN values with 0, which may not be the desired behavior
   print(df)
```

To avoid this error, you should make sure that the values you use to fill in NaN values are appropriate for the data and the analysis you are performing. For example, filling in NaN values with a value of 0 might skew the data and lead to incorrect results.

#getting a bit more general

Yes, dropna() is a good way to clean NaNs from dataframes, but there are other built-in functions in Pandas that can help you clean data more generally. Here are a few examples:



replace(): This function can be used to replace specific values in a dataframe with other values. You can use this to replace values that are not NaN but still indicate missing data, such as dashes or other symbols. For example:

```python
import pandas as pd

# Create a sample dataframe with dashes indicating missing data
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Energy Usage (kWh/m^2/year)': [150, '-', 200, pd.np.nan, 300]}
df = pd.DataFrame(data)

# Replace dashes with NaN values
df.replace('-', pd.np.nan, inplace=True)

# This replaces dashes with NaN values, making it easier to use `dropna()` to remove missing data
print(df)
```

another example of using replace

```python
import pandas as pd

# Create a sample dataframe with similar but not identical values
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'City': ['New York', 'New York City', 'NYC', 'Brooklyn', 'Manhattan']}
df = pd.DataFrame(data)

# Replace similar values with a single value
df['City'] = df['City'].replace(['New York', 'New York City', 'NYC'], 'New York')

# This replaces all similar values with a single value, making it easier to analyze the data
print(df)
```

another example of using replace

```python
import pandas as pd

# Create a sample dataframe with blank values
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Energy Usage (kWh/m^2/year)': [150, '', 200, pd.np.nan, 300]}
df = pd.DataFrame(data)

# Replace blank values with NaN
df.replace('', pd.np.nan, inplace=True)

# This replaces all blank values with NaN, making it easier to use the data for analysis
print(df)
```

In this example, we use the replace() function to replace all blank values ('') in the Energy Usage (kWh/m^2/year) column with NaN (pd.np.nan). The inplace=True argument ensures that the original dataframe is modified rather than a copy being made.

interpolate(): This function can be used to fill in missing values in a dataframe using various interpolation methods. You can use this to fill in missing data that may not be NaN, such as missing sensor readings or data that was not recorded. For example:

```python
import pandas as pd

# Create a sample dataframe with missing values
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Energy Usage (kWh/m^2/year)': [150, pd.np.nan, 200, pd.np.nan, 300]}
df = pd.DataFrame(data)

# Interpolate missing values using a linear method
df.interpolate(method='linear', inplace=True)

# This fills in missing values with interpolated values, making it easier to use the data for analysis
print(df)
```

drop(): This function can be used to drop specific rows or columns from a dataframe based on certain criteria. You can use this to remove data that may be incorrect or irrelevant, such as data that was recorded incorrectly or data from sensors that were malfunctioning. For example:

```python
import pandas as pd

# Create a sample dataframe with irrelevant data
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Energy Usage (kWh/m^2/year)': [150, pd.np.nan, 200, pd.np.nan, 300],
        'Has Solar Panels?': ['Yes', 'No', 'No', 'No', 'Yes']}
df = pd.DataFrame(data)

# Remove data from rows where 'Has Solar Panels?' is 'No'
df = df[df['Has Solar Panels?'] != 'No']

# This removes data that is not relevant to the analysis, making it easier to work with the remaining data
print(df)
```
