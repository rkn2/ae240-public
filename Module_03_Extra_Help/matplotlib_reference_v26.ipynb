# Matplotlib Reference: Implicit vs Explicit Style

When you look up matplotlib examples online you'll see two different styles. **In this course we use the explicit style (`fig, ax`).** This notebook explains both so you can read code you find online and convert it if needed.

## Setup

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 14, 8, 20, 6]
````

---
## Style 1: Implicit (`plt.` style)

You'll see this in many online tutorials and Stack Overflow answers. It's shorter, but has an important limitation in Jupyter notebooks.

````python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Implicit style')
plt.show()
````

### ⚠️ The Gotcha: Split Cells

`pyplot` uses a hidden "state machine" to store the current plot. If you split commands across notebook cells, **the state resets between cells** and you lose your plot.

Run the two cells below to see this happen:

````python
plt.plot(x, y)  # Cell A — draws the line
````

````python
plt.title('My title')  # Cell B — pyplot starts a FRESH plot here; the line is gone
````

To avoid this in the implicit style, all your `plt.` commands must be in **one cell**. The explicit style doesn't have this problem.

---
## Style 2: Explicit (`fig, ax` style) — used in this course

`plt.subplots()` creates and returns the Figure and Axes as variables. Because they're stored, you can modify them across cells without losing work.

````python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Explicit style')
````

````python
# Come back later and add to the same figure — no problem
ax.set_title('Updated title')
fig  # type the variable name to display the figure
````

---
## Side-by-Side Reference

| Action | Implicit (`plt.`) | Explicit (`ax.`) |
|--------|-------------------|------------------|
| Plot data | `plt.plot(x, y)` | `ax.plot(x, y)` |
| X-axis label | `plt.xlabel('text')` | `ax.set_xlabel('text')` |
| Y-axis label | `plt.ylabel('text')` | `ax.set_ylabel('text')` |
| Title | `plt.title('text')` | `ax.set_title('text')` |
| Legend | `plt.legend()` | `ax.legend()` |
| Display figure | *(auto in Jupyter)* | type `fig` |

**Key pattern:** axis methods use `set_` prefix. The plotting call (`ax.plot`) and its keyword arguments (`color`, `marker`, `linestyle`, `label`) are **identical** in both styles.

---
## Converting Online Code to `fig, ax` Style

If you find an example that uses `plt.`:

1. Add `fig, ax = plt.subplots()` at the top
2. `plt.plot(...)` → `ax.plot(...)` *(same arguments)*
3. `plt.xlabel('x')` → `ax.set_xlabel('x')`
4. `plt.ylabel('y')` → `ax.set_ylabel('y')`
5. `plt.title('t')` → `ax.set_title('t')`
6. `plt.legend()` → `ax.legend()` *(no change needed)*
7. Add `fig` on its own line at the end to display

### Example Conversion

**Before (implicit):**
```python
plt.scatter(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('My chart')
plt.legend(['data'])
```

**After (explicit):**
```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('My chart')
ax.legend(['data'])
fig
```

## Multiple Subplots

`plt.subplots(nrows, ncols)` creates a grid of axes. This is where the name "subplots" comes from — for most of this course we use the default `nrows=1, ncols=1`.

````python
fig, axes = plt.subplots(1, 2)  # 1 row, 2 columns

axes[0].plot(x, y)
axes[0].set_title('Left plot')

axes[1].plot(y, x)
axes[1].set_title('Right plot')

fig.tight_layout()  # prevents labels from overlapping
fig
````
