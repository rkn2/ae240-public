# Error messages

Here are some notes on how to debug common errors in Python, along with examples that pertain to architectural engineering:

## Syntax errors: 
These errors occur when the code violates the syntax rules of Python. These are the easiest errors to fix, and the interpreter will usually provide a helpful error message indicating the line and type of error.


```python
# Incorrect syntax - forgot to close parentheses
room_volume = length * width * height
print "The volume of the room is:", room_volume

# Correct syntax
room_volume = length * width * height
print("The volume of the room is:", room_volume)
```

## Name errors: 

These errors occur when a variable or function is used before it is defined. 

```python
# Name error - variable name not defined
room_area = length * width
print("The area of the room is:", room_area)

# Correct code
length = 10
width = 5
room_area = length * width
print("The area of the room is:", room_area)
```

##Type errors: 
These errors occur when an operation is performed on an object of the wrong type.

```python
# Type error - using a string instead of a number
room_height = "2"
room_volume = length * width * room_height
print("The volume of the room is:", room_volume)

# Correct code
room_height = 2
room_volume = length * width * room_height
print("The volume of the room is:", room_volume)
```

## Index errors: 
These errors occur when an index is out of range. 

```python
# Index error - trying to access an element that doesn't exist
room_dimensions = [10, 5]
print("The height of the room is:", room_dimensions[2])

# Correct code
room_dimensions = [10, 5, 2]
print("The height of the room is:", room_dimensions[2])
```

## Logical errors: 

These errors occur when the program runs without any error messages, but produces incorrect results. 

```python
# Logical error - using the wrong formula for calculating wall area
wall_area = 2 * length * height + 2 * width * height
print("The area of the walls is:", wall_area)

# The correct formula for wall area is 2*height*(length + width)
# Correct code
wall_area = 2 * height * (length + width)
print("The area of the walls is:", wall_area)
```

These are just a few examples of how you could apply the concepts of debugging in Python to problems in architectural engineering. By breaking down the problem into smaller parts, testing each part individually, and using techniques like print statements, you can more easily identify and fix errors in your code.
