## Linear Regression with a List: 


### example 1
Suppose you have a list of the heights of five different buildings in meters, and you want to determine if there is a relationship between the height of a building and its year of construction. Write a Python program that uses scipy.stats.linregress() to perform a linear regression on the data and print out the slope, intercept, r-value, p-value, and standard error of the regression.


```python
import numpy as np
from scipy import stats

# Define the list of building heights and years
building_heights = [80, 120, 100, 90, 110] # height of each building in meters
years = [1990, 2000, 1980, 1960, 2010] # year of construction for each building

# Perform a linear regression on the data
slope, intercept, r_value, p_value, std_err = stats.linregress(years, building_heights)

# Print out the results
print("Slope: ", slope)
print("Intercept: ", intercept)
print("R-Value: ", r_value)
print("P-Value: ", p_value)
print("Standard Error: ", std_err)
```

### example 2

Suppose you have a list of the ages of ten different buildings in years, and you want to determine if there is a relationship between the age of a building and the number of floors it has. Write a Python program that uses scipy.stats.linregress() to perform a linear regression on the data and print out the slope, intercept, r-value, p-value, and standard error of the regression.

```python
import numpy as np
from scipy import stats

# Define the list of building ages and number of floors
building_ages = [20, 10, 30, 40, 50, 5, 15, 25, 35, 45] # age of each building in years
num_floors = [3, 2, 4, 5, 6, 1, 3, 2, 4, 5] # number of floors for each building

# Perform a linear regression on the data
slope, intercept, r_value, p_value, std_err = stats.linregress(building_ages, num_floors)

# Print out the results
print("Slope: ", slope)
print("Intercept: ", intercept)
print("R-Value: ", r_value)
print("P-Value: ", p_value)
print("Standard Error: ", std_err)
```

## linear regression with a dataframe

### example 1

Suppose you have a dataframe that contains information about the energy usage and age of different buildings, and you want to determine if there is a relationship between the age of a building and its energy usage. Write a Python program that uses scipy.stats.linregress() to perform a linear regression on the data and print out the slope, intercept, r-value, p-value, and standard error of the regression.

```python
import pandas as pd
from scipy import stats

# Create a sample dataframe with information about building energy usage and age
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Year Built': [1990, 2000, 1980, 1960, 2010],
        'Energy Usage (kWh/m^2/year)': [150, 250, 200, 100, 300]}
df = pd.DataFrame(data)

# Perform a linear regression on the data
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Year Built'], df['Energy Usage (kWh/m^2/year)'])

# Print out the results
print("Slope: ", slope)
print("Intercept: ", intercept)
print("R-Value: ", r_value)
print("P-Value: ", p_value)
print("Standard Error: ", std_err)
```

### example 2

Suppose you have a dataframe that contains information about the energy usage and building size of different buildings, and you want to determine if there is a relationship between the building size and its energy usage. Write a Python program that uses scipy.stats.linregress() to perform a linear regression on the data and print out the slope, intercept, r-value, p-value, and standard error of the regression.

```python
import pandas as pd
from scipy import stats

# Create a sample dataframe with information about building energy usage and size
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Size (m^2)': [1000, 800, 1200, 900, 1500],
        'Energy Usage (kWh/year)': [15000, 12000, 18000, 13500, 22500]}
df = pd.DataFrame(data)

# Perform a linear regression on the data
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Size (m^2)'], df['Energy Usage (kWh/year)'])

# Print out the results
print("Slope: ", slope)
print("Intercept: ", intercept)
print("R-Value: ", r_value)
print("P-Value: ", p_value)
print("Standard Error: ", std_err)
```

## Debugging
Here are some common errors that can occur when using scipy.stats.linregress() to perform a linear regression in Python, and how to debug them:

ValueError: This error occurs when the input data used for the linear regression contains missing or invalid values, such as NaN or infinity. To debug this error, check the data used for the linear regression and make sure that there are no missing or invalid values.

```python
from scipy import stats

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, float('nan'), 5]
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
# Output: ValueError: Input contains NaN, infinity or a value too large for dtype('float64').
```

TypeError: This error occurs when the input data used for the linear regression has the wrong data type, or when the input data has different lengths. To debug this error, check the data used for the linear regression and make sure that it has the correct data type and length.

```python
from scipy import stats

x = [1, 2, 3, 4, 5]
y = [1, 2, 3]
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
# Output: TypeError: unsupported operand type(s) for /: 'float' and 'int'.
```
