# Using the len() function

The len() function in Python is used to find the length of an object. The object can be a string, list, tuple, set, or dictionary. The len() function returns the number of elements in the object.

Here are 10 examples related to architectural engineering that students can use to practice using the len() function in Python:



Find the length of a string that contains the name of a building.

```python
building_name = "Empire State Building"
name_length = len(building_name)
print("The length of the building name is:", name_length)
```

Find the number of windows in a room.

```python
room_windows = ["window1", "window2", "window3"]
num_windows = len(room_windows)
print("The number of windows in the room is:", num_windows)
```

Find the number of doors in a building.

```python
building_doors = {"front": 1, "back": 1, "side1": 2, "side2": 2}
num_doors = len(building_doors)
print("The number of doors in the building is:", num_doors)
```

In the context of dataframes, the len() function can be used to find the number of rows in a dataframe. 

```python
import pandas as pd

# Create a dataframe of building materials
materials = pd.DataFrame({
    'Material': ['Wood', 'Brick', 'Concrete', 'Steel', 'Glass'],
    'Density (kg/m^3)': [700, 1900, 2300, 7850, 2500],
    'Thermal Conductivity (W/mK)': [0.12, 0.69, 1.7, 50, 0.7]
})

# Get the number of rows in the dataframe
num_rows = len(materials)

# Print the number of rows
print("The dataframe contains", num_rows, "rows.")
```

In this example above, we use the len() function to find the number of rows in the materials dataframe. The len() function returns the number of rows in the dataframe, which is then stored in the num_rows variable. We then print out the number of rows using a formatted string. This example can be useful when working with large datasets and you need to quickly determine the size of your data.

Let's try another example with dataframes:

```python
import pandas as pd

# Create a dictionary with information about buildings affected by an earthquake
building_data = {
    'name': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
    'damage': ['minor', 'destroyed', 'major', 'minor', 'destroyed'],
    'height': [50, 80, 120, 100, 60],
    'floors': [10, 15, 20, 18, 12],
    'location': ['Downtown', 'Suburb', 'Downtown', 'Suburb', 'Downtown']
}

# Create a pandas dataframe from the dictionary
building_df = pd.DataFrame(building_data)

# Print the dataframe to the console
print(building_df)
```

We first create a dictionary called building_data that contains information about five different buildings affected by an earthquake, including their names, the extent of the damage they suffered, their height, the number of floors, and their location. We then use the pd.DataFrame() function to create a pandas dataframe called building_df from this dictionary. Finally, we print the dataframe to the console.



We can then use the len() function along with conditional filtering to count the number of buildings that were destroyed in the earthquake. Finally, we print the number of buildings destroyed to the console.

```python
# Get the number of buildings destroyed
num_buildings_destroyed = len(building_df[building_df['damage'] == 'destroyed'])

# Print the number of buildings destroyed
print('The earthquake destroyed', num_buildings_destroyed, 'buildings.')
```

This example demonstrates how the len() function can be used in the context of pandas dataframes to quickly and easily count the number of rows that meet certain criteria. In this case, we used it to count the number of buildings destroyed in an earthquake, but the same technique could be applied to other types of data analysis as well.
