## using standard deviation



Suppose you have a list of the structural stee beam lengths, and you want to remove any that are more than a standard deviation from the mean. Write a Python program that uses the mean() and std() functions in NumPy to identify and remove the outliers from the list.

```python
import numpy as np

# Define the list of steel beam lengths in meters
steel_beam_lengths = [10, 11, 12, 9, 13, 8, 15, 16, 7, 14, 18, 6]

# Calculate the mean and standard deviation of the list
mean = np.mean(steel_beam_lengths)
std = np.std(steel_beam_lengths)

# Filter the list to remove any beam lengths that are more than one standard deviation from the mean
filtered_beam_lengths = [length for length in steel_beam_lengths if abs(length - mean) <= std]

# Print out the filtered list of beam lengths
print(filtered_beam_lengths)
```

In this example, we first define a list of steel beam lengths in meters. We then calculate the mean and standard deviation of the list using the mean() and std() functions from NumPy. We use these values to filter the list of beam lengths to remove any lengths that are more than one standard deviation away from the mean using a list comprehension. Finally, we print out the filtered list of beam lengths.

Suppose you have a dataframe that contains information about the energy usage and age of different buildings, and you want to remove any buildings that have an energy usage that is more than one standard deviation from the mean. Write a Python program that uses the mean() and std() functions in Pandas to identify and remove the outliers from the dataframe.

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage and age
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E', 'Building F'],
        'Year Built': [1990, 2000, 1980, 1960, 2010, 1975],
        'Energy Usage (kWh/m^2/year)': [150, 250, 200, 1000, 300, 400]}
df = pd.DataFrame(data)

# Calculate the mean and standard deviation of the energy usage column
mean = df['Energy Usage (kWh/m^2/year)'].mean()
std = df['Energy Usage (kWh/m^2/year)'].std()

# Filter the dataframe to remove any buildings with energy usage more than 2 standard deviations from the mean
filtered_df = df[abs(df['Energy Usage (kWh/m^2/year)'] - mean) <=  std]

# Print out the filtered dataframe
print(filtered_df)
```

Suppose you have a dataframe that contains information about the energy usage and age of different buildings, and you want to remove any buildings that have an energy usage that is more than three standard deviations from the mean. Write a Python program that uses the mean() and std() functions in Pandas to identify and remove the outliers from the dataframe.

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage and age
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Year Built': [1990, 2000, 1980, 1960, 2010],
        'Energy Usage (kWh/m^2/year)': [150, 250, 200, 1000, 300]}
df = pd.DataFrame(data)

# Calculate the mean and standard deviation of the energy usage column
mean = df['Energy Usage (kWh/m^2/year)'].mean()
std = df['Energy Usage (kWh/m^2/year)'].std()

# Filter the dataframe to remove any buildings with energy usage more than 3 standard deviations from the mean
filtered_df = df[(df['Energy Usage (kWh/m^2/year)'] >= mean - 3 * std) & (df['Energy Usage (kWh/m^2/year)'] <= mean + 3 * std)]

# Print out the filtered dataframe
print(filtered_df)
```

## using a threshold

Suppose you have a dataframe that contains information about the energy usage and size of different buildings, and you want to remove any buildings that have an energy usage more than 12000. Write a Python program to remove any buildings that exceed the threshold.

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage and size
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Size (m^2)': [1000, 800, 1200, 900, 1500],
        'Energy Usage (kWh/year)': [15000, 12000, 18000, 13500, 22500]}
df = pd.DataFrame(data)

# Calculate the threshold for removing outliers
threshold = 12000

# Remove any buildings that exceed the threshold
df = df.loc[df['Energy Usage (kWh/year)'] <= threshold]

# Print the resulting dataframe
print(df)
```

Suppose you have a dataframe that contains information about the energy usage and size of different buildings, as well as the number of floors in each building, and you want to remove any buildings that have an energy usage more than two standard deviations from the mean for each column. Write a Python program to remove any buildings that exceed the threshold for any column.

```python
import pandas as pd

# Create a sample dataframe with information about building energy usage, size, and number of floors
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Size (m^2)': [1000, 800, 1200, 900, 1500],
        'Energy Usage (kWh/year)': [1500000, 12000, 18000, 13500, 22500],
        'Number of Floors': [3, 2, 4, 5, 6]}
df = pd.DataFrame(data)

# Calculate the threshold for removing outliers for each column
threshold_energy = 20000
threshold_size = 1000
threshold_floors = 5

# Remove any buildings that exceed the threshold for any column
df = df.loc[(df['Energy Usage (kWh/year)'] <= threshold_energy) & (df['Size (m^2)'] <= threshold_size) & (df['Number of Floors'] <= threshold_floors)]

# Print the resulting dataframe
print(df)
```

## debugging

Here are some common errors that can occur when trying to remove outliers from a Pandas dataframe using standard deviation, and how to debug them:

Not Handling Missing Values: If the input dataframe contains missing values such as NaN or None, then it can cause errors when trying to calculate the mean or standard deviation. To debug this error, first check for missing values in the dataframe using pandas.DataFrame.isnull() or pandas.DataFrame.isna(), and then either remove the rows or fill in the missing values using pandas.DataFrame.fillna() or pandas.DataFrame.dropna().

```python
import pandas as pd

# Create a sample dataframe with missing values
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Size (m^2)': [1000, 800, None, 900, 1500],
        'Energy Usage (kWh/year)': [15000, 12000, 18000, 13500, 22500]}
df = pd.DataFrame(data)

# Calculate the threshold for removing outliers
threshold = df['Energy Usage (kWh/year)'].mean() + 2 * df['Energy Usage (kWh/year)'].std()
# Output: ValueError: cannot convert float NaN to integer
```

Not Considering the Distribution of the Data: Standard deviation is sensitive to outliers, but it can also be affected by the distribution of the data. For example, if the data has a skewed distribution, then using standard deviation to remove outliers may not be appropriate. To debug this error, plot the data and check the distribution, and consider using other methods such as quartiles or percentiles to remove outliers.

```python
import pandas as pd

# Create a sample dataframe with skewed data
data = {'Building': ['Building A', 'Building B', 'Building C', 'Building D', 'Building E'],
        'Size (m^2)': [1000, 800, 1200, 900, 1500],
        'Energy Usage (kWh/year)': [15000, 12000, 18000, 13500, 35000]}
df = pd.DataFrame(data)

# Calculate the threshold for removing outliers
threshold = df['Energy Usage (kWh/year)'].mean() + 2 * df['Energy Usage (kWh/year)'].std()
# Output: The threshold is too low and does not remove the outlier.
```
